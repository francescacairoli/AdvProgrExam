```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_int = open('performance_different_types_int_1.txt','r')
f_double = open('performance_different_types_double_1.txt','r')
```

```python
number_lookups = []
int_global_time = []
int_time_per_oper = []

for i, line in enumerate(f_int):
    if i > 2:
        n, glob, op, l = line.split()
        number_lookups.append(int(n))
        int_global_time.append(float(glob))
        int_time_per_oper.append(float(op))
        
```

```python
double_global_time = []
double_time_per_oper = []

for i, line in enumerate(f_double):
    if i > 2:
        n, glob, op, l = line.split()
        double_global_time.append(float(glob))
        double_time_per_oper.append(float(op))
```

```python
plt.semilogy(number_lookups, int_time_per_oper, label="int")
plt.semilogy(number_lookups, double_time_per_oper, label="double")
plt.rcParams.update({'font.size': 20})
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
plt.xlabel("n. of lookups")
plt.ylabel("time")
plt.legend(prop={'size': 15})
plt.show()
```
