```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_bal = open('benchmark_balanced.txt','r')
f_unbal = open('benchmark_unbalanced.txt','r')
f_map = open('benchmark_map.txt','r')
f_umap = open('benchmark_unordered_map.txt','r')
```

```python
number_lookups = []
bal_global_time = []
bal_time_per_oper = []
unbal_global_time = []
unbal_time_per_oper = []
map_global_time = []
map_time_per_oper = []
umap_global_time = []
umap_time_per_oper = []


for i, line in enumerate(f_bal):
    if i > 1:
        n, glob, op, l = line.split()
        number_lookups.append(int(n))
        bal_global_time.append(float(glob))
        bal_time_per_oper.append(float(op))
        
for i, line in enumerate(f_unbal):
    if i > 1:
        n, glob, op, l = line.split()
        unbal_global_time.append(float(glob))
        unbal_time_per_oper.append(float(op))
        
for i, line in enumerate(f_map):
    if i > 1:
        n, glob, op, l = line.split()
        map_global_time.append(float(glob))
        map_time_per_oper.append(float(op))
        
for i, line in enumerate(f_umap):
    if i > 1:
        n, glob, op, l = line.split()
        umap_global_time.append(float(glob))
        umap_time_per_oper.append(float(op))
```

```python
plt.semilogy(number_lookups, bal_time_per_oper, label="bal")
plt.semilogy(number_lookups, unbal_time_per_oper, label="unbal")
plt.semilogy(number_lookups, map_time_per_oper, label="map")
plt.semilogy(number_lookups, umap_time_per_oper, label="umap")
plt.xlabel("N")
plt.ylabel("look-ups time")
plt.legend()
plt.show()
```
